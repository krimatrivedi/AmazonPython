```python
# importing libraries
from bs4 import BeautifulSoup
from urllib.request import urlopen

import requests
import json
import re

product_json = {}

def main(URL):

	# opening our output file in append mode
	File = open("out.csv", "a")

	# specifying user agent, You can use other user agents
	# available on the internet
	HEADERS = ({'User-Agent':
		'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36',
						'Accept-Language': 'en-US, en;q=0.5'})

	# Making the HTTP Request
	webpage = requests.get(URL, headers=HEADERS)

	# Creating the Soup Object containing all data
	soup = BeautifulSoup(webpage.content, "lxml")

# This block of code will help extract the Brand of the item
	# retrieving product title
	product_json['The Product Title'] = []
	
	try:
		# Outer Tag Object
		title = soup.find("span",attrs={"id": 'productTitle'})
		# Inner NavigableString Object
		title_value = title.string

		# Title as a string value
		title_string = title_value.strip().replace(',', '')
		
	except AttributeError:
		title_string = "URL not found"
	product_json['The Product Title'].append(title_string)		
	print("product Title = ", title_string)

	# saving the title in the file
	File.write(f"{title_string},")
  #image
	product_json['Product Image URL']=[]
	for link in soup.find_all('img',attrs={'class':'a-dynamic-image image-stretch-horizontal frontImage'}):
    		
				product_json['Product Image URL'].append(link.get('src'))
		
 # retrieving price
	product_json['Price of the Product']=[]
	try:
		price = soup.find("span", attrs={'class': 'a-size-base a-color-price a-color-price'}).string.strip().replace(',', '')
		product_json['Price of the Product'].append(price)	  
	 
    
	except AttributeError:
		price = "NA"
	print("Products price = ", price)

	# saving
	File.write(f"{price},")

	#details
	product_json['Product Details'] =[]
	
	
	for x in soup.find_all('div',attrs={'id':'detailBulletsWrapper_feature_div'}):
		for xx in x.find_all('div',attrs={'id':'detailBullets_feature_div'}):
			for mm in xx.find_all('span',attrs={'class':'a-list-item'}):
				for tag in mm.find_all('span'):
				#print(tag.get_text(),"pro")
					product_json['Product Details'].append(tag.get_text())
				
	#details
	
	
	# closing the file
	File.close()
  		
  


if __name__ == '__main__':
# opening our url file to access URLs
#	file = open("url.txt", "r")
	# iterating over the urls
#	for links in file.readlines():
#		print("yo,",links)
#		main(links)
	# Saving the scraped html file


# Saving the scraped data in json format
	with open("url.txt","r") as urllist,open("product.json", 'w') as outfile:
		for URL in urllist.readlines():
			 main(URL) 
			
		 		
			 json.dump(product_json, outfile, indent=4)
		
			 outfile.write("\n")
```
